```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
x_sqaured = x**2
```

```python
plt.plot(x, x, marker = "*")
plt.plot(x, x_sqaured, marker="o")
plt.title("Comparison Graph")
plt.show()
```

```python
sales_in_2018 = np.array([4,8,2,6,9])
```

```python
plt.bar(x,sales_in_2018, width=0.3)
plt.show()
```
